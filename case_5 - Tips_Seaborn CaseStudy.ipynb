## Tips Dataset Using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head()
```

```python
tips.columns
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
plt.ylabel('Total Bill')
plt.show()
```

```python
sns.countplot(x='smoker',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.boxplot(x='day',y='tip',data=tips)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',hue='sex',data=tips,palette='coolwarm')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',hue='smoker',data=tips,palette='Set1')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',hue='smoker',data=tips,palette='Set1',split=True)
plt.show()
```

```python
sns.violinplot(y='day',x='total_bill',hue='smoker',data=tips,palette='Set1',split=True)
plt.show()
```

```python
sns.violinplot(y='day',x='total_bill',hue='sex',data=tips,palette='Set1',split=True)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',hue='smoker',data=tips,jitter=True)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
fig = plt.figure(figsize = (12,8))
sns.swarmplot(x='day',y='total_bill',hue='smoker',data=tips,size=15)
plt.show()
```
