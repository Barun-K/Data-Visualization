## Movies Dataset Using Matplotlib/Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mv = pd.read_csv('tmdb_5000_movies.csv')
```

```python
mv.head()
```

```python
mv.columns
```

1.Draw a correlogram of movies dataset ignoring the id column.
-----------------------------------------------------------------

```python
mv_c = mv.drop(['id'],axis=1)
```

```python
mv_c.columns
```

```python
mv_c.corr()
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(mv_c.corr(), annot = True, cmap = "rainbow")    
plt.show() 
```

2) .Draw a cluster map of the correlation of all numeric column except the id column. What do you observe
-----------------------

```python
mv.info()
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
mv_num = mv.select_dtypes(include=numerics)
mv_num1=mv_num.drop(['id'],axis=1)
mv_numcl=mv_num1.corr()
```

```python
plt.figure(figsize = (10,8))
sns.clustermap(mv_numcl,annot = True, cmap = "rainbow") 
plt.show() 
```

3). Show the top 10 movies based on pouplarity whose vote_average is between a)4 to 6 b)6 to 8 c)8 to 10
--------------------------------------------------------

```python
# a) 4 to 6
mv1=mv[['original_title','popularity']][(mv.vote_average<6.0) & (mv.vote_average>4.0)].sort_values(by=['popularity'],ascending=False).head(10)
mv1
```

```python
plt.figure(figsize = (10,8))
plt.barh(mv1['original_title'],mv1['popularity'])
```

```python
# b) 6 to 8
mv2=mv[['original_title','popularity']][(mv.vote_average<8.0) & (mv.vote_average>6.0)].sort_values(by=['popularity'],ascending=False).head(10)
mv2
```

```python
plt.figure(figsize = (10,8))
plt.barh(mv2['original_title'],mv2['popularity'])
```

```python
# c) 8 to 10
mv3=mv[['original_title','popularity']][(mv.vote_average<10.0) & (mv.vote_average>8.0)].sort_values(by=['popularity'],ascending=False).head(10)
mv3
```

```python
plt.figure(figsize = (10,8))
plt.barh(mv3['original_title'],mv3['popularity'])
```

4). Show the distribution of each numerical column(except id) when original language is es,de,fr
-----------------------------------------------

```python
mv_n = mv[(mv.original_language == 'es')|(mv.original_language == 'de')|(mv.original_language == 'fr')]
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
mv_num4 = mv_n.select_dtypes(include=numerics)
mv_num4=mv_num4.drop(['id'],axis=1)
mv_num4['original_language'] = mv['original_language']
mv_num4
```

```python
sns.pairplot(mv_num4)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,3,1)
sns.barplot(x='original_language',y='budget',data=mv_num4)
plt.subplot(2,3,2)
sns.barplot(x='original_language',y='popularity',data=mv_num4)
plt.subplot(2,3,3)
sns.barplot(x='original_language',y='revenue',data=mv_num4)
plt.subplot(2,3,4)
sns.barplot(x='original_language',y='runtime',data=mv_num4)
plt.subplot(2,3,5)
sns.barplot(x='original_language',y='vote_average',data=mv_num4)
plt.subplot(2,3,6)
sns.barplot(x='original_language',y='vote_count',data=mv_num4)
plt.show()
```
