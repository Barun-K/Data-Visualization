
## Tips/Iris/Flight Dataset Using Matplotlib/Seaborn

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib as mpl 
```

```python
# datasets available in searborn
sns.get_dataset_names() 
```

```python
tips = sns.load_dataset("tips")  
```

```python
tips.head() 
```

```python
fig = plt.figure(figsize = (12,8)) 
# distribution plot 
sns.distplot(tips["total_bill"]) 
# create edges line in the plot 
mpl.rcParams["patch.force_edgecolor"] = True 
# color the edges of each bar
mpl.rcParams["patch.edgecolor"] = "Green"
# change width of edge line 
mpl.rcParams["patch.linewidth"] = 2
plt.title("Distribution Plot for Total_Bill of Tips dataset") 
plt.savefig("plot59.jpg") 
plt.show() 
```

```python
plt.figure(figsize = (12,8))
# kde (kernel density estimation) removes the curved line  
sns.distplot(tips["tip"], kde = False) 
mpl.rcParams["patch.force_edgecolor"] = True 
mpl.rcParams["patch.edgecolor"] = "Red"
mpl.rcParams["patch.linewidth"] = 3
plt.title("Distribution of Tip") 
plt.xlabel("Tip")
plt.ylabel("Frequency") 
plt.savefig("Plot61.jpg") 
plt.show()
```

```python
# joint plot is for bivariate analysis
plt.figure(figsize = (12,8))
sns.jointplot(x = "total_bill", y = "tip", data = tips) 
# by default plot is scatter 
mpl.rcParams["patch.force_edgecolor"] = True 
mpl.rcParams["patch.edgecolor"] = "Purple"
mpl.rcParams["patch.linewidth"] = 3
plt.title("Distribution of Total_Bill with Tip") 
plt.savefig("Plot62.jpg") 
plt.show() 
```

```python
# using kind plot can be changed to hexgonal or any other form of display 
plt.figure(figsize = (12,8))
sns.jointplot(x = "total_bill", y = "tip", kind = 'hex', data = tips) 
mpl.rcParams["patch.edgecolor"] = "black"
plt.savefig("plot63.jpg") 
plt.show()
```

```python
plt.figure(figsize = (12,8))
# regular expression for total_bill and tip
sns.jointplot(x = "total_bill", y = "tip", kind = 'reg', data = tips) 
mpl.rcParams["patch.edgecolor"] = "yellow"
plt.savefig("plot64.jpg") 
plt.show()
```

```python
plt.figure(figsize = (12,8))
# contour for total_bill and tip 
sns.jointplot(x = "total_bill", y = "tip", kind = 'kde', data = tips) 
plt.savefig("plot65.jpg") 
plt.show() 
```

```python
plt.figure(figsize = (12,8))
sns.pairplot(tips)
plt.savefig("Plot66.jpg") 
plt.show() 
```

```python
plt.figure(figsize = (12,8))
sns.pairplot(tips, hue = 'sex') 
plt.savefig("Plot67.jpg") 
plt.show() 
```

```python
plt.figure(figsize = (12,8)) 
sns.kdeplot(tips['tip']) 
plt.savefig("plot68.jpg")
plt.show() 
```

```python
iris = sns.load_dataset("iris") 
iris.head() 
```

```python
# Grid plot of iris dataset 
plt.figure(figsize = (12,8))
grid = sns.PairGrid(iris, hue = 'species') 
grid.map(plt.scatter).add_legend() 
plt.savefig("plot69.jpg")
plt.show() 
```

```python
# plot scatter plot on the upper side & kernel density estimation plot on the lower side of the diagonal 
grid = sns.PairGrid(iris)
grid.map_diag(plt.hist)
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot) 
plt.savefig("Plot70.jpg") 
plt.show() 
```

```python
plt.figure(figsize = (12,8)) 
grid = sns.FacetGrid(tips, col = "time", row = "smoker") 
grid.map(plt.hist, "total_bill") 
mpl.rcParams["patch.force_edgecolor"] = True 
mpl.rcParams["patch.edgecolor"] = "Green"
mpl.rcParams["patch.linewidth"] = 2
plt.savefig("plot71.jpg")
plt.show() 
```

```python
plt.figure(figsize = (12,8)) 
grid = sns.FacetGrid(tips, col = "time", row = "smoker") 
grid.map(plt.scatter, "total_bill", "tip") 
mpl.rcParams["patch.edgecolor"] = "Red"
plt.savefig("plot73.jpg")
plt.show() 
```

```python
plt.figure(figsize = (12,8)) 
grid = sns.JointGrid(x = "total_bill", y = "tip", data = tips) 
grid.plot(plt.scatter, sns.distplot) 
mpl.rcParams["patch.force_edgecolor"] = True 
plt.savefig("plot72.jpg") 
plt.show() 
```

```python
sns.barplot(x = "sex", y = "total_bill", hue = "time", data = tips, palette = "Greens") 
sns.set_style("ticks") 
plt.savefig("plot74.jpg") 
plt.show() 
```

```python
sns.set_style("darkgrid") 
f, axes = plt.subplots(1, 2, figsize = (12,8)) 
sns.boxplot(y = "total_bill", x = "sex", data = tips, palette = "gist_rainbow_r" , ax = axes[0]) 
sns.boxplot(y = "total_bill", x = "smoker", data = tips, palette = "rocket", ax = axes[1]) 
plt.savefig("plot75.jpg") 
plt.show() 
```
